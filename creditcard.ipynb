```python
x = data.drop(['Amount','Class'],axis = 1)
y = data["Amount"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cutoff = 0.7                              # decide on a cutoff limit
y_pred_classes = np.zeros_like(ypred)    # initialise a matrix full with zeros
y_pred_classes[ypred > cutoff] = 1 
```

```python
y_test_classes = np.zeros_like(ypred)
y_test_classes[ytest > cutoff] = 1
```

```python
confusion_matrix(y_test_classes, y_pred_classes)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test_classes, y_pred_classes)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test_classes, y_pred_classes)
```
